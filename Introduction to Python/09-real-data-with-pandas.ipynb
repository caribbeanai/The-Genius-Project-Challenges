### Load the library and the data

Pandas reads a spreadsheet style file (a CSV) into a table called a DataFrame. The football data lives in the Week 2 folder next door.

```python
import pandas as pd

teams = pd.read_csv("../Week 2/data/teams.csv")
print("Rows (teams):", len(teams))
print("Columns:", len(teams.columns))
teams.head()
```

### Looking at columns

Pick a single column by name. Ask for the biggest or smallest values.

```python
print("The teams in the data:")
print(list(teams["team"]))
print()
print("Highest scoring teams on average:")
print(teams[["team", "avg_goals_scored"]].sort_values("avg_goals_scored", ascending=False).head())
```

### Filtering rows

Keep only the rows that pass a test. Here: teams that average more than 1.8 goals a game.

```python
attackers = teams[teams["avg_goals_scored"] > 1.8]
print("Strong attacking teams:")
print(attackers[["team", "avg_goals_scored", "avg_goals_conceded"]])
```

### Quick summaries

Pandas can describe a whole column in one call.

```python
print("Average goals scored across all teams:", round(teams["avg_goals_scored"].mean(), 2))
print("Best defence (fewest conceded):")
print(teams.loc[teams["avg_goals_conceded"].idxmin(), ["team", "avg_goals_conceded"]])
```
