### Making a dictionary

Each item is a `label: value` pair inside curly braces.

```python
team = {
    "name": "Brazil",
    "ranking": 3,
    "avg_goals": 1.9,
    "titles": 5,
}
print(team["name"], "have won", team["titles"], "World Cups")
```

### Reading and changing values

```python
print("Ranking:", team["ranking"])
team["ranking"] = 1          # they climbed to the top
team["nickname"] = "Selecao" # add a new label
print(team)
```

### A list of dictionaries

This is the shape of almost every real dataset: many rows, each with the same labels. The football tables in the Week 2 folder look exactly like this.

```python
teams = [
    {"name": "Argentina", "avg_goals": 2.0},
    {"name": "Japan",     "avg_goals": 1.3},
    {"name": "Morocco",   "avg_goals": 1.4},
]
for t in teams:
    print(f'{t["name"]} score about {t["avg_goals"]} goals a game')
```
