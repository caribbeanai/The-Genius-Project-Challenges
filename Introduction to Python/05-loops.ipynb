### A for loop

Read it as: *for each team in the group, print the team*.

```python
group = ["Brazil", "Serbia", "Switzerland", "Cameroon"]
for team in group:
    print(team)
```

Notice the indent (the spaces at the start of the line). The indented line is the part that repeats. Python uses indents to see what belongs to the loop.

### Looping with a counter

```python
scores = [2, 0, 3, 1]
running_total = 0
for goals in scores:
    running_total = running_total + goals
    print(f"After this match the total is {running_total}")
```

### Repeating a set number of times

`range(5)` gives the numbers 0, 1, 2, 3, 4.

```python
for matchday in range(1, 4):
    print(f"Matchday {matchday} kicks off")
```
