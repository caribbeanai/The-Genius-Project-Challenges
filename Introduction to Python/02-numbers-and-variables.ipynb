### Making a variable

The `=` sign means *put the value on the right into the name on the left*.

```python
goals = 3
print(goals)
```

### Maths

Python is a calculator. It knows `+`, `-`, `*` (times) and `/` (divide).

```python
home_goals = 2
away_goals = 1
total_goals = home_goals + away_goals
print("Total goals in the match:", total_goals)
```

### Numbers with decimals

Whole numbers are called integers. Numbers with a dot, like `1.5`, are called floats. Football averages are usually floats.

```python
goals_per_game = 2.4
matches = 5
expected_total = goals_per_game * matches
print("Over 5 matches we expect about", expected_total, "goals")
```

### Variables can change

You can update a box with a new value at any time.

```python
score = 0
score = score + 1     # a goal is scored
score = score + 1     # another goal
print("Current score:", score)
```
