### Joining strings

The easiest way to build a message is an f-string. Put an `f` before the quotes, then drop variables inside curly braces `{ }`.

```python
team = "Brazil"
goals = 3
print(f"{team} scored {goals} goals")
```

### Useful string tricks

```python
team = "argentina"
print(team.upper())        # ARGENTINA
print(team.capitalize())   # Argentina
print(len(team))           # how many letters
```
