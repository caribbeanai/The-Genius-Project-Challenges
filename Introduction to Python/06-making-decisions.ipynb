### if and else

Python checks the question after `if`. If it is true, it runs the indented block. Otherwise it runs the `else` block.

```python
home_goals = 2
away_goals = 1

if home_goals > away_goals:
    print("Home team wins")
else:
    print("Home team did not win")
```

### Comparisons

These questions give back either `True` or `False`.

`>` greater than, `<` less than, `==` equal to, `!=` not equal to.

```python
a = 3
b = 3
print(a == b)   # True
print(a != b)   # False
print(a > 5)    # False
```

### Three way choice with elif

A football match has three results, so we need three paths.

```python
home_goals = 1
away_goals = 1

if home_goals > away_goals:
    print("Home win")
elif home_goals == away_goals:
    print("Draw")
else:
    print("Away win")
```

### Deciding inside a loop

This is exactly what a simple predictor does: look at each match and call it.

```python
matches = [(2, 0), (1, 1), (0, 3)]   # (home goals, away goals)
for home, away in matches:
    if home > away:
        print(f"{home}-{away}: home win")
    elif home == away:
        print(f"{home}-{away}: draw")
    else:
        print(f"{home}-{away}: away win")
```
