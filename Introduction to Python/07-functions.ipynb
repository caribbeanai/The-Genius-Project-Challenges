### Defining a function

`def` starts a function. The words in brackets are the inputs.

```python
def match_result(home_goals, away_goals):
    if home_goals > away_goals:
        return "home win"
    elif home_goals == away_goals:
        return "draw"
    else:
        return "away win"

# Now use it as many times as we like.
print(match_result(2, 0))
print(match_result(1, 1))
print(match_result(0, 2))
```

### Returning a number

`return` sends a value back so you can store or print it.

```python
def points_from(result):
    if result == "win":
        return 3
    elif result == "draw":
        return 1
    return 0

total = points_from("win") + points_from("draw") + points_from("win")
print("Points from three matches:", total)
```

### A small prediction function

Here is a tiny predictor. It is only a rule, but it already looks like the models you will build in Week 2.

```python
def who_is_favourite(rating_a, rating_b):
    if rating_a > rating_b:
        return "Team A"
    elif rating_a < rating_b:
        return "Team B"
    return "Too close to call"

print(who_is_favourite(88, 72))
print(who_is_favourite(70, 90))
```
