### Making a list

Use square brackets and separate items with commas.

```python
group = ["Brazil", "Serbia", "Switzerland", "Cameroon"]
print(group)
print("Teams in this group:", len(group))
```

### Reading items

Python counts from 0. The first item is number 0, the second is number 1, and so on.

```python
print("First team:", group[0])
print("Second team:", group[1])
print("Last team:", group[-1])   # -1 means the last one
```

### Changing a list

```python
group.append("added team")   # add to the end
print(group)
group[0] = "Argentina"       # replace the first item
print(group)
```

### A list of numbers

Python has quick tools for number lists.

```python
scores = [2, 0, 3, 1, 2]
print("Total goals:", sum(scores))
print("Best single score:", max(scores))
print("Average score:", sum(scores) / len(scores))
```
